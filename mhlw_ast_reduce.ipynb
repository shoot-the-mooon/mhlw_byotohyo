```python
#csv の読み込み
df_shisetsu = pd.read_csv("/content/drive/MyDrive/mhlw_data/施設票_令和4年.csv")
```

#セル結合されたカラムの処理



```python
#一番使えそうなカラム名を含む行番号
base_column_index = 3
base_col = df_shisetsu.iloc[base_column_index]
```

```python
#base_columのうち欠損する箇所を上の行から補完
for i in range(base_column_index, 0,  -1):
  df_shisetsu.iloc[base_column_index].fillna(df_shisetsu.iloc[i - 1])
```


#重複カラムの一時的な処理

```python
#重複するカラム名のリストを作成
df_dup = base_col.value_counts()
df_dup = df_dup[df_dup>1]
df_dup
```

```python
#シリーズをリストに
dup_col_list = df_dup.index.tolist()

#dup_col_listをイテレートして添字付加
for  column in dup_col_list:
  # columnの値を持つ列のインデックスのリストを取得
  indices = base_col[base_col == column].index

  # インデックスを使って値を更新
  for i, col in enumerate(indices, start=1):
      df_shisetsu.at[base_column_index, col] = f'{column}_{i}'

# 更新された行を表示
df_shisetsu.iloc[base_column_index]
```

```python
#上部の不要な行削除(後でエクセルに再度結合)とインデックスの振り直し
df_shisetsu.columns = df_shisetsu.iloc[base_column_index]
#必須項目とかの行も削除
df_shisetsu.drop(df_shisetsu.index[:base_column_index+2], inplace=True)
df_shisetsu.reset_index(drop=True, inplace=True)
df_shisetsu.head()
```

#*の数を確認


```python
#＊があるカラムとその行数を確認
search_string = '*'

counts = {}
for column in df_shisetsu.columns:
    count = df_shisetsu[column].apply(lambda x: x == search_string).sum()
    if count > 0:
      counts[column] = count

# 結果を表示
for column, count in counts.items():
    print(f"{column}: {count}")
```

#月間列を合計して年間列に反映

```python
key = '救急車の受入件数（令和3年4月）'  #始まりの列名
target = '救急車の受入件数'               #合計を反映する列
stride = 12                              #keyから何番目まで合計するか(年間合計なら12)

columns = df_shisetsu.columns.tolist()

#指定した12個の列の合計を別のデータフレームに代入
try:
  key_idx = columns.index(key)
  #ポップアップで確認
  col_str = '\n'.join([col for col in columns[key_idx:key_idx+stride]])
  res = input(f'合計するカラム名 : \n{col_str}\n[y/n]')
  if res == 'y':
    #keyカラムから12列を*→5、intに変換、合計を指定した列に上書き
    df_shisetsu.iloc[:, key_idx:key_idx+stride] = df_shisetsu.iloc[:, key_idx:key_idx+stride].replace('*', '5').astype(int)
    df_shisetsu.loc[:, target] = df_shisetsu.iloc[:, key_idx:key_idx+stride].sum(axis=1)
    print(f'__________________________\n{df_shisetsu[target]}')
  else:
    print("canceled")
except ValueError:
  print("key not found")
```

#関数にして他のカラムも処理

```python
def sum_columns(key, target, df=df_shisetsu, stride=12):

  """
  指定したkeyカラムからstride数右側の範囲で*→５変換
  合計をtargetカラムに上書き

  parameters
  ____________
  key :始まりの列名
  target : 合計を反映する列
  stride : keyから右何列目まで合計するか(年間合計なら12)
  df : データフレーム

  Return
  ______________
  df[target] : 合計結果

  Example
  ______________
  sum_columns('休日に受診した患者延べ数（令和2年4月)', '休日に受診した患者延べ数')

  """
  columns = df.columns.tolist()

  try:
    key_idx = columns.index(key)
    #ポップアップで確認
    col_str = '\n'.join([col for col in columns[key_idx:key_idx+stride]])
    res = input(f'合計するカラム名 : \n{col_str}\n[y/n]')
    if res == 'y':
      #keyカラムから12列を*→5、intに変換、合計を指定した列に上書き
      df.iloc[:, key_idx:key_idx+stride] = df.iloc[:, key_idx:key_idx+stride].replace('*', '5').astype(int)
      df.loc[:, target] = df.iloc[:, key_idx:key_idx+stride].sum(axis=1)
      print(f'__________________________\n{df[target]}')
    else:
      print("canceled")
  except ValueError:
    print("Error : key not found")
```

```python
#「*の数を確認」セクションの出力をもとに手動で関数に引数を渡していく
sum_columns('休日に受診した患者延べ数（令和3年4月)', '休日に受診した患者延べ数')
```

```python
sum_columns('うち診察後直ちに入院となった患者延べ数（令和3年4月）_1', 'うち診察後直ちに入院となった患者延べ数_1')
```

```python
sum_columns('うち診察後直ちに入院となった患者延べ数（令和3年4月）_2', 'うち診察後直ちに入院となった患者延べ数_2')
```

```python
sum_columns('夜間・時間外に受診した患者延べ数（令和3年4月)', '夜間・時間外に受診した患者延べ数')

#日本語の）と英語の）でkeyerrorが出ている
```

#データ型を直してxlsx出力

```python
#intに変換可能な列のみint変換
for col in df_shisetsu.columns:
  temp = pd.to_numeric(df_shisetsu[col], errors='coerce')
  if not temp.isnull().any():
    try:
      df_shisetsu[col] = df_shisetsu[col].astype(int)
    except: #不動小数点の場合はfloatに
      df_shisetsu[col] = df_shisetsu[col].astype(float)
```

#様式２の読み込みとカラム処理

```python
#カラム処理、整形の関数を定義

def topdown_col(df_byoto):

  #base_columのうち欠損する箇所を上の行から補完
  for i in range(base_column_index, 0,  -1):
    df_byoto.iloc[base_column_index] = df_byoto.iloc[base_column_index].fillna(df_byoto.iloc[i-1])
  for col in df_byoto.columns: #カラムも補完に使用
    if pd.isna(df_byoto.at[base_column_index, col]):
      df_byoto.at[base_column_index, col] = col
  df_byoto.head()

def dup_col(df_byoto):
  #重複するカラム名のリストを作成
  df_dup = base_col.value_counts()
  df_dup = df_dup[df_dup>1]
  print(' 重複するカラムのリスト')
  print(df_dup)

    #シリーズをリストに
  dup_col_list = df_dup.index.tolist()

  #dup_col_listをイテレートして添字付加
  for  column in dup_col_list:
    # columnの値を持つ列のインデックスのリストを取得
    indices = base_col[base_col == column].index

    # インデックスを使って値を更新
    for i, col in enumerate(indices, start=1):
      df_byoto.at[base_column_index, col] = f'{column}_{i}'

  # 更新された行を表示
  print('__________________\n添字処理後の全体カラム')
  print(df_byoto.iloc[base_column_index])

def del_col(df_byoto):
  #上部の不要な行削除(後でエクセルに再度結合)とインデックスの振り直し
  df_byoto.columns = df_byoto.iloc[base_column_index]
  #必須項目とかの行も削除
  df_byoto.drop(df_byoto.index[:base_column_index+3], inplace=True)
  df_byoto.reset_index(drop=True, inplace=True)
  print('__________________\n削除後の全体カラム')
  print(df_byoto.head())

def astype_df(df_byoto):
  #intに変換可能な列のみint変換
  for col in df_byoto.columns:
    temp = pd.to_numeric(df_byoto[col], errors='coerce')
    if not temp.isnull().any():
      try: #この段階では整数と浮動小数が含まれる
        df_byoto[col] = df_byoto[col].astype(int)
      except: #不動小数点の場合はfloatに
        df_byoto[col] = df_byoto[col].astype(float)

def count_ast(df_byoto):
  #＊があるカラムとその行数を確認
  search_string = '*'

  counts = {}
  for column in df_byoto.columns:
      count = df_byoto[column].apply(lambda x: x == search_string).sum()
      if count > 0:
        counts[column] = count

  return counts
```
